**Lo que sigue es una función**

```python
def mostrar_nombre(nombre, edad):

  if nombre[-1] == 'a':
    pronombre = 'Ella'
  else:
    pronombre = 'El'

  edad_persona = f'{pronombre} es {nombre} y tiene {edad} años'
  return edad_persona

imprimir_funcion = mostrar_nombre(nombre, edad)
print(imprimir_funcion)
```
